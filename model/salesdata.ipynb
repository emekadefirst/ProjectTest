```python
import re
import pandas as pd
```

```python
data = pd.read_csv("dataset.csv")
```

```python
data.head()
data.info()
data.describe()
```

```python
data.isnull().sum()
df = data.dropna(how='all')
df.drop_duplicates()
```

```python
df['Country'] = df['Country'].str.replace(r'[Xx]', '', regex=True)
df['Description'] = df['Description'].str.replace(r'[\^ö\$]', '', regex=True)
df['StockCode'] = df['StockCode'].str.replace(r'[\^\ö]', '', regex=True)
df['Quantity'] = df['Quantity'].str.replace(r'[@]', '', regex=True)
```

```python
df['CustomerID'] = df['CustomerID'].str.replace(r'[# &]', '', regex=True)
df['UnitPrice'] = df['UnitPrice'].str.replace(r'[Ww]', '', regex=True)
emoji_pattern = re.compile(
    "["
    u"\U0001F600-\U0001F64F"  
    u"\U0001F300-\U0001F5FF"  
    u"\U0001F680-\U0001F6FF"  
    u"\U0001F1E0-\U0001F1FF"  
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    "]+",
    flags=re.UNICODE,
)

df['Country'] = df['Country'].str.replace(emoji_pattern, '', regex=True)
df['Country'] = df['Country'].str.replace(r'[Xx]', '', regex=True)
df['Country'] = df['Country'].str.replace(r'[Y]', '', regex=True)
df['InvoiceNo'] = df['InvoiceNo'].str.replace(r'[ä]', '', regex=True)
```
